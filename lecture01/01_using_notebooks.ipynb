## Skill 3 — Variables persist across cells

Once a cell defines a variable, every later cell can use it. This is why notebooks feel like a chain — but it also causes the most common beginner bug (next skill).

Run the two demo cells in order. The second one uses values from the first without redefining them.

````python
# DEMO part 1 — define the kinetic parameters of an enzyme (e.g. hexokinase)
Vmax = 120.0    # µmol / min / mg enzyme
Km   = 0.15     # mM glucose
print(f"Defined Vmax = {Vmax} and Km = {Km}")
````

````python
# DEMO part 2 — Michaelis–Menten velocity at one substrate concentration
substrate = 0.5   # mM glucose
v = Vmax * substrate / (Km + substrate)
print(f"At [S] = {substrate} mM,  v = {v:.1f} µmol/min/mg")
````

## Skill 4 — When you get `NameError`, check the run order

Variables only exist *after* the cell that creates them has been run. If you skip a cell — or run cells out of order — Python won't find the name and you'll see:

```
NameError: name 'leaf_area_cm2' is not defined
```

**This is the #1 thing that confuses beginners.** The fix is almost never to change code — it's to run the earlier cell first.

````python
# DEMO — run THIS cell first, and you'll get a NameError. That's intentional.
# A plant physiologist measured leaf area and chlorophyll content.
specific_chlorophyll = chlorophyll_ug / leaf_area_cm2
print(f"Specific chlorophyll: {specific_chlorophyll:.2f} µg/cm²")
````

````python
# This cell defines the missing variables. Run it, THEN re-run the cell above.
leaf_area_cm2 = 38.5
chlorophyll_ug = 412.0
````

## Skill 5 — Getting help on a function

You don't need to remember every function. Two tricks let you ask Python what something does, right inside the notebook:

- Type a function name followed by `?`  &nbsp;→&nbsp; e.g. `round?`
- Or call `help(name)`  &nbsp;→&nbsp; e.g. `help(round)`

Both open the official documentation in a panel below the cell.

````python
# DEMO — ask Python what 'round' does (handy for reporting pH to 2 decimals, etc.)
round?
````
