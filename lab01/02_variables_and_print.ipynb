## Skill 3 — f-strings (formatted strings)

Comma-separated `print()` is fine, but f-strings give you full control. Two rules:

1. Put `f` immediately before the opening quote: `f"..."`.
2. Anything in `{ }` inside the string is replaced by the value of that variable.

You can also format the number: `{value:.2f}` rounds to 2 decimal places, `{count:,}` adds thousands separators.

````python
# DEMO — formatted report for a NaCl stock solution
solute = "NaCl"
molarity = 0.15378     # mol/L
volume_ml = 500

print(f"Prepared {volume_ml} mL of {molarity:.3f} M {solute}")
````
