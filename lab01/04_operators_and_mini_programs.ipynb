## Skill 1 — Arithmetic operators

| Operator | Meaning | Example |
|----------|---------|---------|
| `+ - * /` | usual arithmetic | `5 / 2 → 2.5` |
| `**` | power | `2 ** 10 → 1024` |
| `//` | floor division | `7 // 2 → 3` |
| `%` | remainder (modulo) | `7 % 2 → 1` |

`**` is especially useful in biology — anything that doubles, decays or grows exponentially uses it.

```python
# DEMO — serial dilution. Each step dilutes 1:10, so after n steps the dilution is 10**n.
initial_conc_ug_ml = 1000
n_steps = 4
final_conc = initial_conc_ug_ml / (10 ** n_steps)
print(f"After {n_steps} ×10 dilutions: {final_conc} µg/mL")
```

## Skill 3 — Putting it together: a molarity calculator

Most useful programs are just a handful of variables, a calculation, and a formatted print statement. Here's a complete mini-program that calculates the molarity of a NaCl solution.

**Formula:** $\text{molarity (M)} = \dfrac{\text{mass (g)} / \text{MW (g/mol)}}{\text{volume (L)}}$

```python
# DEMO — molarity of an NaCl stock
solute = "NaCl"
mass_g = 5.844
molecular_weight = 58.44     # g/mol
volume_ml = 100

volume_l = volume_ml / 1000
moles = mass_g / molecular_weight
molarity = moles / volume_l

print(f"{mass_g} g {solute} in {volume_ml} mL  →  {molarity:.3f} M")
```
