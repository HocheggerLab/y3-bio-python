## Skill 4 — `bool` and inspecting types with `type()`

A bool has only two possible values: `True` or `False` (capitalised, no quotes). They're how Python answers yes/no questions.

If you're ever unsure what type something is, call `type(value)`.

````python
# DEMO — quality control flags for a sample
is_contaminated = False
passed_qc = True
print(is_contaminated, passed_qc)
print(type(is_contaminated))
````
