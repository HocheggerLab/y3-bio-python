# Introduction to Matplotlib: Figure and Axes

In this notebook, we'll explore the fundamental concepts of matplotlib's object-oriented interface, focusing on the `Figure` and `Axes` objects. Understanding these concepts is crucial for creating and customizing plots effectively.

We'll use simple dummy data and follow a pattern of showing examples, then giving you practice exercises.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Understanding Figure and Axes

In matplotlib:
- **Figure**: The entire window or page that contains everything (like a canvas)
- **Axes**: The actual plot area where data is displayed (despite the name, it's the whole plot, not just the axis!)

Think of it like this:
- Figure = the entire canvas/paper
- Axes = the drawing area within that canvas

### Example 1.1: Creating a Figure and Axes

```python
# Create a figure and a single axes
fig, ax = plt.subplots()

# Create some dummy data y=2x
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot on the axes
ax.plot(x, y)

```

### Example 1.2: Adding Labels and Title

```python
# Create figure and axes
fig, ax = plt.subplots()

# Dummy data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot the data
ax.plot(x, y)

# Add labels and title
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.set_title('My First Plot')

plt.show()
```

## 2. Customizing Plot Appearance

You can customize line colors, styles, and markers.

### Example 2.1: Line Colors and Styles

```python
fig, ax = plt.subplots()

x = [1, 2, 3, 4, 5]
y = [1, 4, 2, 3, 5]

# Plot with red color and dashed line
ax.plot(x, y, color='red', linestyle='--')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('Plot with Red Dashed Line')

plt.show()
```

### Example 2.2: Adding Markers

```python
fig, ax = plt.subplots()

x = [1, 2, 3, 4, 5]
y = [1, 4, 2, 3, 5]

# Plot with markers at data points
ax.plot(x, y, color='blue', linestyle='-', marker='o', markersize=8)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('Plot with Markers')

plt.show()
```

### Example 2.3: Common Line Styles and Markers

Here are some common options:

**Line styles:**
- `'-'`: solid line
- `'--'`: dashed line
- `'-.'`: dash-dot line
- `':'`: dotted line

**Markers:**
- `'o'`: circle
- `'s'`: square
- `'^'`: triangle up
- `'*'`: star
- `'+'`: plus

```python
fig, ax = plt.subplots()

x = np.linspace(0, 10, 20)
y = x ** 2

# Plot with green color, dotted line, and star markers
ax.plot(x, y, color='green', linestyle=':', marker='*', markersize=10)

ax.set_xlabel('X')
ax.set_ylabel('X squared')
ax.set_title('Customized Plot Style')

plt.show()
```

## 3. Multiple Lines on One Plot

You can plot multiple datasets on the same axes.

### Example 3.1: Plotting Multiple Lines

```python
fig, ax = plt.subplots()

x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]  # Linear
y2 = [1, 4, 9, 16, 25]  # Quadratic

# Plot both lines
ax.plot(x, y1, label='Linear')
ax.plot(x, y2, label='Quadratic')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Multiple Lines')
ax.legend()  # Show legend

plt.show()
```

### Example 3.2: Multiple Lines with Different Styles

```python
fig, ax = plt.subplots()

x = np.linspace(0, 5, 20)
y1 = x
y2 = x ** 2
y3 = x ** 3

# Plot with different colors and styles
ax.plot(x, y1, color='blue', linestyle='-', marker='o', label='Linear')
ax.plot(x, y2, color='red', linestyle='--', marker='s', label='Quadratic')
ax.plot(x, y3, color='green', linestyle='-.', marker='^', label='Cubic')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Comparison of Functions')
ax.legend()

plt.show()
```

## 4. Figure Size and Resolution

You can control the size of your figure.

### Example 4.1: Setting Figure Size

```python
# Create a figure with specified size (width, height) in inches
fig, ax = plt.subplots(figsize=(10, 6))

x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y)
ax.set_xlabel('X')
ax.set_ylabel('sin(X)')
ax.set_title('Larger Figure Size')

plt.show()
```

## 5. Multiple Subplots

One of the most powerful features is creating multiple plots in a grid layout.

### Example 5.1: Two Subplots Side by Side

```python
# Create 1 row, 2 columns of subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 2, 3, 5]
y2 = [5, 3, 4, 2, 1]

# Plot on first axes
ax1.plot(x, y1, color='blue')
ax1.set_title('Plot 1')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

# Plot on second axes
ax2.plot(x, y2, color='red')
ax2.set_title('Plot 2')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

plt.tight_layout()  # Adjust spacing to prevent overlap
plt.show()
```

### Example 5.2: Grid of Subplots

```python
# Create 2 rows, 2 columns of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

x = np.linspace(0, 10, 100)

# Top-left plot (axes[0, 0])
axes[0, 0].plot(x, x)
axes[0, 0].set_title('Linear')

# Top-right plot (axes[0, 1])
axes[0, 1].plot(x, x**2)
axes[0, 1].set_title('Quadratic')

# Bottom-left plot (axes[1, 0])
axes[1, 0].plot(x, x**3)
axes[1, 0].set_title('Cubic')

# Bottom-right plot (axes[1, 1])
axes[1, 1].plot(x, np.sqrt(x))
axes[1, 1].set_title('Square Root')

plt.tight_layout()
plt.show()
```

## 6. Other Common Plot Types

Beyond line plots, matplotlib supports many other plot types.

### Example 6.1: Scatter Plot

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)

ax.scatter(x, y, color='purple', alpha=0.6, s=100)  # s is marker size
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.set_title('Scatter Plot')

plt.show()
```

### Example 6.2: Bar Plot

```python
fig, ax = plt.subplots(figsize=(8, 6))

categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]

ax.bar(categories, values, color='skyblue')
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_title('Bar Plot')

plt.show()
```

### Example 6.3: Histogram

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Generate random data from normal distribution
data = np.random.randn(1000)

ax.hist(data, bins=30, color='coral', alpha=0.7, edgecolor='black')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title('Histogram')

plt.show()
```
